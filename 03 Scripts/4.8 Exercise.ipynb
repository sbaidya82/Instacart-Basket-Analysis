# 06. Creating the spending flag for each user based on the average price accross all their orders

```python
# Creating Average_price column
ords_prods_merge['Average_price']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Creating spender_flag 
ords_prods_merge.loc[ords_prods_merge['Average_price']<10,'spender_flag']='Low spender'
ords_prods_merge.loc[ords_prods_merge['Average_price']>=10,'spender_flag']='High spender'
```

```python
# counts of the spenders
ords_prods_merge['spender_flag'].value_counts(dropna=False)
```

# 07. Create an order frequency flag to find out the frequent and non frequent customers

```python
# Creating median_order column
ords_prods_merge['median_order']=ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Creating order_frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_order']>20,'order_frequency_flag']='Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_order']>10)&(ords_prods_merge['median_order']<20),'order_frequency_flag']='Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_order']<=10,'order_frequency_flag']='Frequent customer'
```

```python
# count of order frequency
ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```
