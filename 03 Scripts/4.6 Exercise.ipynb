# Merging DataFrame

# Evaluation
There is a key column "product_id"  among the two dataframes where we can see that the merging is possible, but which type of the merge should be applied , that depends on the information we expect to see in the final dataframe. Since the information wouldn't be much useful if it doesn't exist in both dataframes, therefore, the inner merge will be a suitable option. 

```python
# test merge without overwriting with merge flag "match"
pd.merge(orders_products_combined,df_prods,on=['product_id'], how='inner', indicator='match')
```

# Evaluation
 As the previous exercise of merged dataframe has created'merge' column as the flag which would be no use for the new merge of 'orders_products_combined' and 'df_prods', for better presentation, the '_merge' column from the previous merge sould be dropped after the new merge has been carried out. 

```python
# actual merge of 'orders_products_combined' and 'df_prods'
ords_prods_merge= orders_products_combined.merge(df_prods, on=['product_id'], indicator='match')
```

```python
# Drop the flag column '_merge' from the previous merge
ords_prods_merge.drop('_merge', axis=1)
```
