# Observation :Explanatory analysis (Products DataFrame)
After the explanatory analysis of the products dataframe it was discovered that there are outliers present in the data.
The prices column has a mean of 9.99 and Standard deviation of 453.6 . This standered deviation is more than 2 standered deviation away from the mean, that indicating the presence of the outlier in the data.
Identifying all data in the prices column that is 2 standared deviation away from the mean resulted in two entries or rows, prices for 14900 and prices for 99999.
These were identified as outliers and were subcequently replaced with the median value of 7.1.
Given that the minimum value of 1  and maximum value of 99999 produce a wide range of values , it is recomended to replace outliers  with the median value.
The median is less sensitive to outliers compared to the mean,making it more appropriate measure for this extreme value.
After replacing the outliers with the median value the mean of the dataframe decreased to 7.6 .
